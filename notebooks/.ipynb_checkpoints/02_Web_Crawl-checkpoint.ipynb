# Introduction

Grabs the web text for a given url and saves it in jsonl format

# Imports

```python
import requests
from bs4 import BeautifulSoup
import jsonlines
from tld import get_tld
```

```python
# Constanst & Functions

url = 'https://www.bayer.com/en/environmental-and-climate-protection.aspx'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.9; rv:45.0) Gecko/20100101 Firefox/45.0'}

def download(url):
    try:
        response = requests.get(url, headers=headers) #  send request
        if response.status_code == 200:
            pass
    except requests.exceptions.ConnectionError:
        print('Something wrong ')
        return False
    soup = BeautifulSoup(response.content, 'html.parser')  #  parse web page
    return soup.getText()
```

```python
text = download(url)
print(text)
```

```python

info = get_tld(url, as_object=True)
print("Domain Name: ", info.domain)
```

```python
# format text as jsonl
# {"text": "Terrible customer service.", "labels": ["negative"], "meta": {"wikiPageID": 1}}
meta = {url: url, 'other': 'foo', 'another': 'bar'} 
company_name = './data/' + get_tld(url, as_object=True).domain +  '.jsonl'
with jsonlines.open(company_name, mode='w') as writer:
    writer.write({'text':text, 'meta': meta})
```
